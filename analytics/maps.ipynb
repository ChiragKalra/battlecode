```python
import pandas as pd
import numpy as np
```

```python
maps = pd.read_csv('maps.csv')
maps
```

```python
maps.columns
```

```python
maps['width'] = maps[' dim'].str.split('*', expand=True)[0]
```

```python
maps['height'] = maps[' dim'].str.split('*', expand=True)[1]
```

```python
maps.head()
```

```python
maps.drop(columns=('width', 'height'))
```

```python
maps['mucks in grid'] = (maps['width'].astype('int64')/5).apply(np.ceil) * (maps['height'].astype('int64')/5).apply(np.ceil)
```

```python
maps.head(17)
```

```python
maps = maps.drop(columns=[' dim',('width', 'height'), 'passability','ind'] )
```

```python
maps
```

```python
maps['avg mucks/ec'] = maps['mucks in grid']/(maps[' default'].astype('int64') + maps[' neutral'].astype('int64')/2)
```

```python
maps
```

```python
maps.to_csv('maps.csv')
```
